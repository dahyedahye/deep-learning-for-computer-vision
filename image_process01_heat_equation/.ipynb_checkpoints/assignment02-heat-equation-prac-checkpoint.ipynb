```python
def create_padded_image(image):
    h = image.shape[0]
    w = image.shape[1]
    
    col_zero = np.zeros(shape = (h,1))
    row_zero = np.zeros(shape = (1,w+2))
    plane_top = np.hstack((col_zero, image))
    plane_top = np.hstack((plane_top, col_zero))
    plane_top = np.vstack((plane_top, row_zero))
    plane_top = np.vstack((plane_top, row_zero))

    plane_bottom = np.hstack((col_zero, image))
    plane_bottom = np.hstack((plane_bottom, col_zero))
    plane_bottom = np.vstack((row_zero, plane_bottom))
    plane_bottom = np.vstack((row_zero, plane_bottom))

    col_zero_for_side = np.zeros(shape = (h+2,1))
    row_zero_for_side = np.zeros(shape = (1,w))

    plane_left = np.vstack((row_zero_for_side, image))
    plane_left = np.vstack((plane_left, row_zero_for_side))
    plane_left = np.hstack((plane_left, col_zero_for_side))
    plane_left = np.hstack((plane_left, col_zero_for_side))

    plane_right = np.vstack((row_zero_for_side, image))
    plane_right = np.vstack((plane_right, row_zero_for_side))
    plane_right = np.hstack((col_zero_for_side, plane_right))
    plane_right = np.hstack((col_zero_for_side, plane_right))

    img_padded = np.vstack((row_zero_for_side, image))
    img_padded = np.vstack((img_padded, row_zero_for_side))
    img_padded = np.hstack((col_zero_for_side, img_padded))
    img_padded = np.hstack((img_padded, col_zero_for_side))
    
    return plane_top, plane_bottom, plane_left, plane_right, img_padded


def heat_equation(image, max_iter, plot_iter_list):
    img = image
    for i in range(max_iter):
        plane_top, plane_bottom, plane_left, plane_right, img_padded = create_padded_image(img)
        lu = plane_top + plane_bottom + plane_left + plane_right - (4*img_padded)
        img_next = img_padded + (1/4*1/4*lu)
        img = img_next[1:120, 1:101]
        
        if i + 1 in plot_iter_list:
            plt.imshow(img, cmap = "gray")
            plt.show()
            print("iter:", i + 1)
    
```
