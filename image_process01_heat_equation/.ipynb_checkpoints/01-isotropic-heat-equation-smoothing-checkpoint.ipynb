## Create padded images to efficiently compute the heat equation  
* Create 4 new images padded at top, bottom, left and right respectively
* Copy over the most top row / bottom row /left column /right column to use their values as padded values at each boundary, following the Neumann boundary condition.

```python
def create_padded_image(image):
    h = image.shape[0]
    w = image.shape[1]
    
    col_zero = np.zeros(shape = (h,1))
    row_zero = np.zeros(shape = (1,w+2))
    
    ## top
    plane_top = np.hstack((col_zero, image))
    plane_top = np.hstack((plane_top, col_zero))
    # pad on the bottom
    pad_bottom_row = plane_top[-1]
    np.reshape(pad_bottom_row, (1,w+2))
    plane_top = np.vstack((plane_top, pad_bottom_row))
    plane_top = np.vstack((plane_top, row_zero))

    ## bottom
    plane_bottom = np.hstack((col_zero, image))
    plane_bottom = np.hstack((plane_bottom, col_zero))
    # pad on the top
    pad_top_row = plane_bottom[0]
    np.reshape(pad_top_row, (1,w+2))
    plane_bottom = np.vstack((pad_top_row, plane_bottom))
    plane_bottom = np.vstack((row_zero, plane_bottom))

    col_zero_for_side = np.zeros(shape = (h+2,1))
    row_zero_for_side = np.zeros(shape = (1,w))

    ## left
    plane_left = np.vstack((row_zero_for_side, image))
    plane_left = np.vstack((plane_left, row_zero_for_side))
    # pad on the right
    pad_right_col = plane_left[:,-1]
    pad_right_col = np.reshape(pad_right_col, (h+2, 1))
    plane_left = np.hstack((plane_left, pad_right_col))
    plane_left = np.hstack((plane_left, col_zero_for_side))

    ## right
    plane_right = np.vstack((row_zero_for_side, image))
    plane_right = np.vstack((plane_right, row_zero_for_side))
    # pad on the left
    pad_left_col = plane_right[:,0]
    pad_left_col = np.reshape(pad_left_col, (h+2, 1))
    plane_right = np.hstack((pad_left_col, plane_right))
    plane_right = np.hstack((col_zero_for_side, plane_right))

    img_padded = np.vstack((row_zero_for_side, image))
    img_padded = np.vstack((img_padded, row_zero_for_side))
    img_padded = np.hstack((col_zero_for_side, img_padded))
    img_padded = np.hstack((img_padded, col_zero_for_side))
    
    return plane_top, plane_bottom, plane_left, plane_right, img_padded
```

## Define a function to compute heat equation

* $\frac{\partial^{2}u}{\partial x^2} = \frac{u(x+\Delta x, y) - 2u(x,y) + u(x-\Delta x,y)}{\Delta x^{2}}$
* $\frac{\partial^{2}u}{\partial y^2} = \frac{u(x, y + \Delta y) - 2u(x,y) + u(x, y - \Delta y)}{\Delta y^{2}}$
* $\Delta u = \frac{u(x+\Delta x, y) + u(x-\Delta x,y) + u(x, y + \Delta y)+u(x, y - \Delta y) - 4u(x,y)}{\Delta h^{2}}$
  * where, $\Delta x = \Delta y = \Delta h$
* $\frac{u^{(t+1)}-u^{(t)}}{\Delta t} = \alpha \cdot \Delta u^{(t)}$
  * Therefore, $u^{(t+1)}(x,y) = u^{(t)}(x,y) + \Delta t \cdot \alpha \cdot \Delta u^{(t)}$  

```python
def heat_equation(image, max_iter, plot_iter_list):
    img = image
    for i in range(max_iter):
        plane_top, plane_bottom, plane_left, plane_right, img_padded = create_padded_image(img)
        lu = plane_top + plane_bottom + plane_left + plane_right - (4*img_padded)
        img_next = img_padded + (DEL_T*ALPHA*lu)
        img = img_next[1:120, 1:101]
        
        if i + 1 in plot_iter_list:
            plt.imshow(img, cmap = "gray")
            plt.show()
            print("iter:", i + 1)
```

## Execute and plot results at iteration 0, 1, 2, 4, 8, 16, 32, 64, 128

```python
# Resize the image to reduce the image size
r = 100.0 / img_gray.shape[1]
dim = (100, int(img_gray.shape[0]*r))
img_small = cv2.resize(img_gray, dim, interpolation = cv2.INTER_AREA)
plot_iter_list = [0, 1, 2, 4, 8, 16, 32, 64, 128]
plt.imshow(img_small, cmap = "gray")
plt.show()
print("iter: 0 - Original image", img_small.shape)
heat_equation(img_small, 128, plot_iter_list)
```
